### 03) WAP to find tuples which have all elements divisible by K from a list of tuples.
**Sample Input:**
```
tuple_list = [(2, 4, 6), (3, 6, 9), (4, 8, 12), (5, 10, 15)]
K = 2

```
**Sample Output:**
```
Tuples divisible by 2 : [(2, 4, 6), (4, 8, 12)]
```

````python
length = int(input('Enter length:'))
l1 = []
for i in range(length):
    l1.append(tuple(map(int, input('Enter Elements:').split())))

k = int(input('Enter k:'))
result = []

for i in l1:
    for j in i:
        if j % k != 0:
            break
    else:
        result.append(i)
print(f'Tuples divisible by 2: {result}')
````

### 11) WAP to rotate the elements of a tuple to the right by `k` positions.

T = (10, 20, 30, 40, 50)

```
(40, 50, 10, 20, 30)
```

### 14) WAP to create (index, value) pairs for prime numbers.

**Sample Input:**
```
T = (4, 7, 9, 11, 15)
```
**Sample Output:**
```
((1, 7), (3, 11))
```

````python
T = tuple(map(int, input('Enter elements: ').split()))
result = tuple()

for i in range(len(T)):
    num = T[i]
    
    if num > 1:
        is_prime = True
        
        for j in range(2, int(num**0.5) + 1):
            if num % j == 0:
                is_prime = False
                break
        
        if is_prime:
            result += ((i, num),)

print(result)
````

### 16) WAP to compute column-wise sum from a tuple of lists.

**Sample Input:**
```
T = ([1, 2, 3], [4, 5, 6], [7, 8, 9])
```

**Sample Output:**
```
12 15 18

```

````python
n = int(input("Enter number of lists: "))
T = []

for i in range(n):
    lst = list(map(int, input(f"Enter elements: ").split()))
    T.append(lst)

T = tuple(T)

column_sums = [sum(col) for col in zip(*T)]

print("Column sums:", ' '.join(map(str, column_sums)))
````
