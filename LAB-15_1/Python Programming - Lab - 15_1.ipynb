### 03) WAP to create Circle class with area and perimeter function to find area and perimeter of circle.

```python
import math
class Circle:
    def __init__(self, radius):
        self.radius = radius
        
    def area(self):
        return math.pi * self.radius ** 2
        
    def perimeter(self):
        return 2 * math.pi * self.radius

c1 = Circle(5)
print(f"Radius: {c1.radius}")
print(f"Area: {c1.area():.2f}")
print(f"Perimeter: {c1.perimeter():.2f}")
```

### 04) Create a class for employees that includes attributes such as name, age, salary, and methods to update and display employee information.

### 05) Create a bank account class with methods to deposit, withdraw, and check balance.

```python
class Bank:
    def __init__(self, balance=0):
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited: {amount}")
        else:
            print("Invalid deposit amount")
    
    def withdraw(self, amount):
        if amount <= 0:
            print("Invalid withdrawal amount")
        elif self.balance >= amount:
            self.balance -= amount
            print(f"Withdrawn: {amount}")
        else:
            print("Insufficient Balance")

    def check_balance(self):
        print(f"Current Balance: {self.balance}")

acc = Bank(5000)

acc.deposit(2000)
acc.withdraw(3000)
acc.withdraw(5000) # show insufficient
acc.check_balance()        
```

### 06) Create a class for managing inventory that includes attributes such as item name, price, quantity, and methods to add, remove, and update items.

```python
class Inventory:
    def __init__(self):
        self.items = {}

    def add(self, name, price, quantity):
        if price < 0 or quantity < 0:
            print("Invalid price or quantity")
            return
            
        self.items[name] = {"price":price, "quantity":quantity}
        print(f"{name} added successfully")

    def remove(self, name):
        if name in self.items:
            del self.items[name]
            print(f"{name} removed")
        else:   
            print(f"{name} doesn't exist")

    def update(self, name, price, quantity):
        if name in self.items:
            if price < 0 or quantity < 0:
                print("Invalid price or quantity")
                return
                
            self.items[name] = {"price": price, "quantity": quantity}
            print(f"{name} updated")
        else:
            print(f"{name} doesn't exist")

    def display(self):
        print("--- Inventory ---")
        for name, details in self.items.items():
            print(f"Item: {name}, Price: {details['price']}, Quantity: {details['quantity']}")

inv = Inventory()

inv.add("Laptop", 50000, 10)
inv.add("Mouse", 500, 50)

inv.display()

inv.update("Mouse", 600, 40)
inv.remove("Laptop")

inv.display()
```

### 09) Define Time class with hour and minute as data member. Also define addition method to add two time objects.

```python
class Time:
    def __init__(self, hour, minute):
        self.hour = hour
        self.minute = minute

    def add(self, t2):
        total_minutes = self.minute + t2.minute
        total_hours = self.hour + t2.hour

        if total_minutes >= 60:
            total_hours += total_minutes // 60
            total_minutes = total_minutes % 60

        return Time(total_hours, total_minutes) # Create new object

    def display(self):
        print(f"{self.hour} hour {self.minute} minutes")

t1 = Time(2, 50)
t2 = Time(1, 30)

t3 = t1.add(t2)

print("Time 1:")
t1.display()

print("Time 2:")
t2.display()

print("After Addition:")
t3.display()
```
