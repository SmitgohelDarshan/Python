### 01) WAP to check whether the given number is positive or negative.

### 05) WAP to check whether the given year is leap year or not.
#### [If a year can be divisible by 4 but not divisible by 100 then it is leap year but if it is divisible by 400 then it is leap year]

```python
year = int(input('Enter the Year:'))

if((year % 4 == 0 or year % 100 != 0) and (year % 400 == 0)):
    print(f"{year} is a Leap year.")
else:
    print(f"{year} is not a Leap year.")
```

### 06) WAP in python to display the name of the day according to the number given by the user.

### 08) WAP to read marks of five subjects. Calculate percentage and print class accordingly. 
#### Fail below 35 
#### Pass Class between 35 to 45 
#### Second Class
#### between 45 to 60
#### First Class between 60 to 70
#### Distinction if more than 70

```python
subject_1 = int(input('Enter marks of Subject 1:'))
subject_2 = int(input('Enter marks of Subject 2:'))
subject_3 = int(input('Enter marks of Subject 3:'))
subject_4 = int(input('Enter marks of Subject 4:'))
subject_5 = int(input('Enter marks of Subject 5:'))

percentage = ((subject_1 + subject_2 + subject_3 + subject_4 + subject_5) / 500) * 100
if(percentage > 70):
    print('Distinction')
elif(percentage > 60):
    print('First Class')
elif(percentage > 45):
    print('Second Class')
elif(percentage > 35):
    print('Pass class')
else:
    print('Fail')
```

### 10) WAP to calculate electricity bill based on following criteria. Which takes the unit from the user.
#### a. First 1 to 50 units – Rs. 2.60/unit
#### b. Next 50 to 100 units – Rs. 3.25/unit
#### c. Next 100 to 200 units – Rs. 5.26/unit
#### d. above 200 units – Rs. 8.45/unit

```python
units = int(input('Enter Units:'))

if(units > 200):
    print(f'{(50 * 2.60) + (50 * 3.25) + (100 * 5.26) + (units - 200) * 8.45}')
elif(units > 100):
    print(f'{(50 * 2.60) + (50 * 3.25) + (units - 100) * 5.26}')
elif(units > 50):
    print(f'{(50 * 2.60) + (units - 50) * 3.25}')
else:
    print(f'{units * 2.60}')
```
