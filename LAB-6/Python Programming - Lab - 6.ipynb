### 09) WAP to count unique items in a list.

### 12) WAP to count occurrences of an element in a list.

```python
l1 = list(map(int, input('Enter a list:').split()))
seen = [] 
for i in l1:
    if not i in seen:
        count = l1.count(i)
        print(f"{i}({count})")
        seen.append(i)      
```

### 14) WAP to extract elements with frequency greater than K.

```python
l1 = list(map(int, input('Enter a list:').split()))
k = int(input('Enter k:'))

l2 = []
seen = []
for i in l1:
    if not i in seen:
        if l1.count(i) > k:
            l2.append(i)
    seen.append(i)  

print(f'List: {l2}')        
```

### 15) WAP to create a list of squared numbers from 0 to 9 with and without using List Comprehension.

```python
l1 = []
for i in range(0,10):
    l1.append(i*i)

print(f'List: {l1}')

l2 = [i*i for i in range(0,10)]
print(f'List: {l2}')
```

### 17) WAP to create a list of common elements from given two lists.

```python
l1 = list(map(int, input('Enter a list 1:').split()))
l2 = list(map(int, input('Enter a list 2:').split()))
l1.sort()
l2.sort()
result = []
i = 0; j = 0

while i < len(l1) and j < len(l2):
    if l1[i] == l2[j]:
        result.append(l1[i])
        i += 1; j += 1
    elif l1[i] > l2[j]:
        j += 1
    else:
        i += 1        


print(f'Common Elements from two List: {result}')        
```
