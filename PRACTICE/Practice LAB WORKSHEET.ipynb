## Practical 1: Student Marks Analyzer

### Problem Definition
Store student data as a list of tuples. Remove duplicate marks, calculate average marks and store result in a dictionary.

### Sample Input
```python
students = [
    (1, "Amit", [78, 85, 90, 85]),
    (2, "Neha", [88, 92, 79, 88]),
    (3, "Ravi", [70, 75, 80, 75])
]
```

### Sample Output
```
{1: ('Amit', 84.33), 2: ('Neha', 86.33), 3: ('Ravi', 75.0)}

```

````python
n = int(input('Enter number of students: '))
students = []

for i in range(n):
    roll = int(input(f'Enter roll number of student {i+1}: '))
    name = input(f'Enter name of student {i+1}: ')
    marks = list(map(int, input(f'Enter marks (space-separated) for {name}: ').split()))

    students.append((roll,name,marks))

result = {}
for roll, name, marks in students:
    unique_marks = list(set(marks))
    avg = sum(unique_marks) / len(unique_marks)
    result[roll] = (name, round(avg, 2))

print('Result', result)
````

## Practical 2: Inventory Management

### Problem Definition
Remove out-of-stock items and calculate total inventory value.

### Sample Input
```python
inventory = {
    'pen': (10, 50),
    'book': (100, 0),
    'pencil': (5, 100)
}
```

### Sample Output
```
{'pen': (10, 50), 'pencil': (5, 100)}
Total Value: 1000
```

````python
n = int(input('Enter number of items: '))
inventory = {}

for i in range(n):
    item = input(f'Enter item {i+1} name: ')
    price = float(input(f'Enter price for {item}: '))
    quantity = float(input(f'Enter quantity for {item}: '))

    inventory[item] = (price, quantity)

filtered_inventory = {}
total_value = 0

for item, (price,qty) in inventory.items():
    if qty > 0:
        filtered_inventory[item] = (price, qty)
        total_value += price * qty

# items_to_remove = []
# total_value = 0

# # First calculate total value and identify items to remove
# for item, (price, quantity) in inventory.items():
#     if quantity == 0:
#         items_to_remove.append(item)
#     else:
#         total_value += price * quantity

# # Remove out-of-stock items
# for item in items_to_remove:
#     del inventory[item]


print('Available Inventory: ', filtered_inventory)
print('Total Inventory Value: ', total_value)
````

## Practical 3: Email Domain Extractor

### Problem Definition
Extract email domains and count occurrences.

### Sample Input
```python
emails = ['a@gmail.com', 'b@yahoo.com', 'c@gmail.com']
```

### Sample Output
```
{'gmail.com': 2, 'yahoo.com': 1}
```

````python
emails = list(map(str,input('Enter emails (space-seperated): ').split()))

domain_count = {}

for email in emails:
    domain = email.split('@')[1]

    # domain_count[domain] = domain_count.get(domain, 0) + 1
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

print(domain_count)
````

## Practical 4: Shopping Cart Billing

### Problem Definition
Calculate item-wise cost and total bill.

### Sample Input
```
cart = [('apple',2), ('banana',3)]
prices = {'apple':50, 'banana':10}
```

### Sample Output
```
{'apple': 100, 'banana': 30}
Total Bill: 130
```

````python
cart = []
n = int(input("Enter number of items in cart: "))
for i in range(n):
    item = input(f"Enter item {i+1} name: ")
    quantity = int(input(f"Enter quantity for {item}: "))
    cart.append((item, quantity))

prices = {}
m = int(input("\nEnter number of items with prices: "))
for i in range(m):
    item = input(f"Enter item {i+1} name: ")
    price = int(input(f"Enter price for {item}: "))
    prices[item] = price

item_costs = {}
total_bill = 0

for item, quantity in cart:
    if item in prices:
        cost = quantity * prices[item]
        item_costs[item] = cost
        total_bill += cost
    else:
        print(f"Warning: No price found for {item}")


# item_costs = {item: quantity * prices[item] for item, quantity in cart}

# # Calculate total bill
# total_bill = sum(item_costs.values())

print("\nItem-wise costs:", item_costs)
print(f"Total Bill: {total_bill}")
````

## Practical 5: Character Position Index

### Problem Definition
Store positions of each character in a string.

### Sample Input
```python
s = "banana"
```

### Sample Output
```
{'b': [0], 'a': [1, 3, 5], 'n': [2, 4]}
```

````python
s = input('Enter string: ')

char_positions = {}

for i, char in enumerate(s):
    if char in char_positions:
        char_positions[char].append(i)
    else:
        char_positions[char] = [i]

print(char_positions)    
````

## Practical 6: Survey Data Aggregation

### Problem Definition
Count responses city-wise and choice-wise.

### Sample Input
```python
responses = [
    ('Amit','Rajkot','Yes'),
    ('Neha','Rajkot','No'),
    ('Ravi','Ahmedabad','Yes')
]
```

### Sample Output
```
{'Rajkot': {'Yes': 1, 'No': 1}, 'Ahmedabad': {'Yes': 1}}
```

````python
n = int(input('Enter number of responses: '))
responses = []

for i in range(n):
    name  = input(f'Enter name {i+1}: ')
    city  = input(f'Enter city for {name}: ')
    choice  = input(f'Enter choice (yes/No) {name}: ')

    responses.append((name,city,choice))

result = {}

for name, city, choice in responses:
    if city not in result:
        result[city] = {}

    if choice not in result[city]:
        result[city][choice] = 0

    result[city][choice] += 1

    # city_dict[choice] = city_dict.get(choice, 0) + 1
    
print(result)    
````
